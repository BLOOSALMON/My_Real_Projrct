```python
# 이진 분류 모델
# sequential : 각 층 순차적 누적
# 딥러닝 모델 구조

# 은닉층 : 훈련용, 2개 시 심층 신경망 (dense)

def create_model():
  model = Sequential()
  model.add(Dense(256, input_shape=(x_train.shape[1],)))  # 입력 특성의 shape은 x_train.shape로부터 자동 유추, node 256 개
  model.add(Dropout(0.3)) # 과적함 방지위해 30% 뉴련 랜덤 제거
  model.add(Dense(256, activation="relu")) # ReLU: 활성화 함수, 비선형성 부여
  model.add(Dense(128, activation="relu"))
  model.add(Dropout(0.3))
  model.add(Dense(64, activation="relu"))
  model.add(Dropout(0.3))

  # output neuron 이진 분류
  model.add(Dense(1, activation = "sigmoid"))

  # 손실함수
  model.compile(loss = "binary_crossentropy", metrics = ["accuracy"], optimizer="adam") # 이진 분류에서 사용하는 손실 함수,  # 정확도를 기준으로 평가, # 대표적인 최적화 알고리즘

  model.summary()
  return model




```

```python
from functools import wraps
import time
def how_long(func):
  def wrapper(*args, **kwargs):
    start_time = time.perf_counter()
    result = func(*args, **kwargs)
    end_time = time.perf_counter()

    total_time = round(end_time - start_time, 3)
    print(f"{func.__qualname__}:{total_time} 초", flush=True)
    return result
  return wrapper
```


. 개념: PDF란?
Probability Density Function (확률 밀도 함수) 는:

연속형 변수의 분포를 나타내는 함수입니다.

예측 확률처럼 0.0 ~ 1.0 사이의 실수 값을 가지는 경우,
그 값들이 어떻게 분포돼 있는지를 시각화하면 모델의 신뢰도와 구분도를 이해할 수 있습니다



딥러닝 예측 확률 PDF를 보는 이유
| ----------------- | ------------------------------- |
| 모델이 확신을 가지고 예측하는지 | 예: 대부분 `0.95` 또는 `0.05` → 확신 있음 |
| 불확실한 예측의 분포       | 예: `0.45 ~ 0.55`에 몰려 있으면 애매한 판단 |
| 클래스 간 구분이 잘 되는지   | 예: 클래스별 예측 확률 분포가 확연히 다르면 좋은 모델 |


```python
def plot_pdf(y_pred, y_test, name=None, smooth=500):
  positives = y_pred[y_test.label == 1]
  negatives = y_pred[y_test.label == 0]
  N = positives.shape[0] # N: 양성 샘플의 수 (예: 534개)
  n = 10
  s = positives # 예측 확률 배열 positives를 s에 복사 (그대로 사용)
  p, x = np.histogram(s, bins=n) # p: 각 구간(bin)의 카운트 값 (길이 = 10) # x: bin의 경계 값 (길이 = 11)
  # ex) p = [12, 34, 78, ..., 1], x = [0.1, 0.2, ..., 1.0]

  x = x[:-1] + (x[1] - x[0])/2 # 중앙값 바꿔 스플라인(부드러운 곡선에 쓸 리스트 생성)

  f = UnivariateSpline(x, p, s = n) # s=n: 스무딩 정도. 작으면 곡선이 더 정확히 맞고, 크면 더 부드러워짐
  plt.plot(x, f(x))

  N = negatives.shape[0]
  n = 10
  s = negatives
  p, x = np.histogram(s, bins = n)

  x = x[:-1] + (x[1] - x[0]) / 2

  f = UnivariateSpline(x, p, s=n)
  plt.plot(x, f(x))
  plt.xlim([0.0, 1.0])
  plt.xlabel('Predicted Probability')
  plt.ylabel('density')
  plt.title('PDF-{}'.format(name))
  plt.show()



```
